```python
import hashlib

# === 1. Робота з текстом ===
def word_counter(text):
    words = text.lower().split()
    word_freq = {}
    for word in words:
        word = word.strip(".,!?\"'():;-")
        if word:
            word_freq[word] = word_freq.get(word, 0) + 1
    return word_freq

# === 2. Інвентаризація продуктів ===
inventory = {
    'яблука': 10,
    'груші': 4,
    'банани': 2,
    'апельсини': 7
}

def update_inventory(product, amount):
    inventory[product] = inventory.get(product, 0) + amount
    if inventory[product] < 0:
        inventory[product] = 0

# === 3. Статистика продажів ===
sales = [
    {"продукт": "ноутбук", "кількість": 2, "ціна": 700},
    {"продукт": "телефон", "кількість": 5, "ціна": 300},
    {"продукт": "монітор", "кількість": 4, "ціна": 250},
    {"продукт": "телефон", "кількість": 1, "ціна": 300},
    {"продукт": "ноутбук", "кількість": 1, "ціна": 700},
]

def calculate_income(sales_list):
    income = {}
    for sale in sales_list:
        product = sale["продукт"]
        amount = sale["кількість"] * sale["ціна"]
        income[product] = income.get(product, 0) + amount
    return income

# === 4. Система управління задачами ===
tasks = {
    "Підготувати звіт": "в процесі",
    "Зустріч з командою": "виконано",
    "Оновити сайт": "очікує"
}

def add_task(name, status):
    tasks[name] = status

def delete_task(name):
    if name in tasks:
        del tasks[name]

def update_task_status(name, new_status):
    if name in tasks:
        tasks[name] = new_status

# === 5. Аутентифікація користувачів ===
users = {
    "admin": {
        "password": hashlib.md5("1234".encode()).hexdigest(),
        "full_name": "Павлов Володимир Володимирович"
    }
}

def authenticate_user(login):
    password = input("Введіть пароль: ")
    hashed_input = hashlib.md5(password.encode()).hexdigest()
    user = users.get(login)
    if user and user["password"] == hashed_input:
        print(f"Вітаємо, {user['full_name']}")
    else:
        print("Невірний логін або пароль.")

# === Приклади використання ===

# 1. Робота з текстом
text = "Привіт привіт як справи? Як справи у тебе? У тебе все добре?"
freq = word_counter(text)
print("Частота слів:", freq)
more_than_3 = [word for word, count in freq.items() if count > 3]
print("Слова, що зустрічаються більше 3 разів:", more_than_3)

# 2. Інвентаризація продуктів
update_inventory("груші", -2)
update_inventory("ківі", 3)
print("Інвентаризація:", inventory)
low_stock = [item for item, count in inventory.items() if count < 5]
print("Товари з кількістю менше ніж 5:", low_stock)

# 3. Статистика продажів
income = calculate_income(sales)
print("Загальний дохід по продуктах:", income)
high_income = [product for product, amount in income.items() if amount > 1000]
print("Продукти з доходом більше 1000:", high_income)

# 4. Задачі
add_task("Написати звіт", "очікує")
update_task_status("Оновити сайт", "в процесі")
delete_task("Зустріч з командою")
print("Список задач:", tasks)
pending = [task for task, status in tasks.items() if status == "очікує"]
print("Задачі зі статусом 'очікує':", pending)

# 5. Аутентифікація
print("Перевірка логіна admin:")
authenticate_user("admin")
```
